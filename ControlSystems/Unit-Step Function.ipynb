```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x,tau):
    return 1-np.exp(-x/tau)
```

```python
tau=1
timesteps=200

x=np.linspace(0,6,timesteps)
y=f(x,tau)
step=np.ones(timesteps)
step[0]=0
```

```python
plt.plot(x,step)
plt.plot(x,y)
plt.legend(["step input","first-order output"])
plt.title("Unit step response of an RC filter")
plt.xlabel("t/tau")
```
