```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants
g = -9.8 #m/s^2
l = 1 #m
m = 1 #kg
b = 0.1 # damping coefficient

# Define state-space model
A = np.array([[0, 1], [g/l, -b/(m*l*l)]])
B = np.array([[0], [1/(m*l*l)]])
C = np.array([[1, 0]])
D = np.array([[0]])
x = np.array([[0.1], [0]])
t = 0.001
u = 0

# Simulate system and plot output
timesteps = 10000
y = np.zeros((timesteps, 1))
for i in range(timesteps):
    y[i] = np.dot(C, x) + np.dot(D, u)
    x_dot = np.dot(A, x) + np.dot(B, u)
    x = x + x_dot*t

plt.plot(y)
plt.xlabel('Time (ms)')
plt.ylabel('Output')
plt.show()
```
