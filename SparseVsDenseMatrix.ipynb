```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse import csr_matrix
```

```python
N = 10000
np.random.seed(99)
A = np.random.binomial(1,0.1,N*N).reshape(N,N)
```

```python
plt.spy(A)
plt.title("Sparse Matrix")
```

```python
np.random.seed(99)
B = np.random.binomial(1,0.9,N*N).reshape(N,N)
```

```python
plt.spy(B)
plt.title("Dense Matrix")
```

```python
A_sparse = csr_matrix(A)
B_sparse = csr_matrix(B)
```

```python
plt.spy(A_sparse)
```
