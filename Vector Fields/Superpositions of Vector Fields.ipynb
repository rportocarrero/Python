```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lims = [-10,10]
dim = 10
x,y = np.meshgrid(np.linspace(lims[0],lims[1],dim),np.linspace(lims[0],lims[1],dim))
```

```python
#uniform flow vector field
def uniform_flow(x,y,v):
    return v,0

u,v = uniform_flow(x,y,10)
plt.quiver(x,y,u,v)
plt.show()
```

```python
#source vector field
def source(x,y,v):
    return v*x/(x**2+y**2),v*y/(x**2+y**2)

u,v = source(x,y,10)
plt.quiver(x,y,u,v)
plt.show()
```

```python
#sink vector field
def sink(x,y,v):
    return -v*x/(x**2+y**2),-v*y/(x**2+y**2)

u,v = sink(x,y,10)
plt.quiver(x,y,u,v)
plt.show()
```

```python
# uniform field superimposed with source
def superpos1(x,y,v,w):
    return w+(v*x)/(x**2+y**2),(v*y)/(x**2+y**2)

u,v = superpos1(x,y,20,5)
plt.streamplot(x,y,u,v)
plt.show()
```

```python
#uniform field superimposed with sink
def superpos2(x,y,v,w):
    return w+(v*-x)/(x**2+y**2),(v*-y)/(x**2+y**2)

u,v = superpos2(x,y,20,5)
plt.streamplot(x,y,u,v)
plt.show()
```

```python
# adding up individual fields
u1,v1 = uniform_flow(x,y,5)
u2,v2 = source(x,y,20)
u,v, = u1+u2,v1+v2

plt.streamplot(x,y,u,v)
plt.show()
```

```python
# magnetic dipole
u1,v1=source(x-5,y,10)
u2,v2=sink(x+5,y,10)
u,v=u1+u2,v1+v2

plt.streamplot(x,y,u,v)
plt.show()
```

```python
#Rotated magnetic dipole
u1,v1=source(x-0.5,y+2.5,2)
u2,v2=sink(x+0.5,y-2.5,2)
u,v=u1+u2,v1+v2

plt.streamplot(x,y,u,v)
plt.show()
```

```python
def uniform(x,y,v):
    return 0,v
```

```python
u1,v1=source(x-0.5,y+2.5,20)
u2,v2=sink(x+0.5,y-2.5,20)
u3,v3=uniform(x,y,1)
u,v=u1+u2+u3,v1+v2+v3

plt.streamplot(x,y,u,v)
plt.show()
```
