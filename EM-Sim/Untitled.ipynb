```python
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
# build grid
x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)
X, Y = np.meshgrid(x,y)
```

```python
# calculate Vectors
Ex = (X + 1)/((X+1)**2 + Y**2) - (X - 1)/((X-1)**2 + Y**2)
Ey = Y/((X+1)**2 + Y**2) - Y/((X-1)**2 + Y**2)
```

```python
# Plot fields
plt.figure(figsize=(10, 10))
plt.streamplot(X,Y,Ex,Ey)
plt.plot(-1,0,'-or')
plt.plot(1,0,'-og')
plt.title('Electromagnetic Field')
```
